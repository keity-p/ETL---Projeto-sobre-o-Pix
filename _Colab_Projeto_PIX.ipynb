```python
# PIX - TRANSFORMANDO MESES EM TRIMESTRES

df_pix['mes'] = df_pix['mes'].replace(1,1)
df_pix['mes'] = df_pix['mes'].replace(2,1)
df_pix['mes'] = df_pix['mes'].replace(3,1)

df_pix['mes'] = df_pix['mes'].replace(4,2)
df_pix['mes'] = df_pix['mes'].replace(5,2)
df_pix['mes'] = df_pix['mes'].replace(6,2)

df_pix['mes'] = df_pix['mes'].replace(7,3)
df_pix['mes'] = df_pix['mes'].replace(8,3)
df_pix['mes'] = df_pix['mes'].replace(9,3)

df_pix['mes'] = df_pix['mes'].replace(10,4)
df_pix['mes'] = df_pix['mes'].replace(11,4)
df_pix['mes'] = df_pix['mes'].replace(12,4)


# VERIFICAÇÃO
df_pix.head(3)
```

* **CRIAÇÃO DE ANÁLISE DE CRESCIMENTO PERCENTUAL DO VOLUME DO PIX AO ANO**

```python
# Selecionando apenas os anos 2021, 2022 e 2023
df_pix_selecionado = df_pix[df_pix['ano'].isin([2021, 2022, 2023])]

# Agrupando por ano e somando as colunas VL_PagadorPF e VL_PagadorPJ
soma_por_ano = df_pix_selecionado.groupby('ano')[['VL_PagadorPF', 'VL_PagadorPJ']].sum()

# Verificação se aparecem os 3 anos (2021, 2022, 2023)
print(soma_por_ano)

# Selecionando os totais para os anos 2021 a 2023
total_2021 = soma_por_ano.loc[2021, 'VL_PagadorPF'] + soma_por_ano.loc[2021, 'VL_PagadorPJ']
total_2022 = soma_por_ano.loc[2022, 'VL_PagadorPF'] + soma_por_ano.loc[2022, 'VL_PagadorPJ']
total_2023 = soma_por_ano.loc[2023, 'VL_PagadorPF'] + soma_por_ano.loc[2023, 'VL_PagadorPJ']

# Aumento percentual entre 2021 e 2023
aumento_percent_1 = ((total_2023 - total_2021) / total_2021) * 100

# Aumento percentual entre 2021 e 2022
aumento_percent_2 = ((total_2022 - total_2021) / total_2021) * 100

# Aumento percentual entre 2022 e 2023
aumento_percent_3 = ((total_2023 - total_2022) / total_2022) * 100

print(f'\nAumento percentual entre 2021 e 2022: {aumento_percent_2:.2f}%')
print(f'Aumento percentual entre 2022 e 2023: {aumento_percent_3:.2f}%')
print(f'\nAumento percentual entre 2021 e 2023: {aumento_percent_1:.2f}%')
```
